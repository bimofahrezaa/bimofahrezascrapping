```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get('https://www.forbes.com/lists/soccer-valuations/?sh=71520119198b')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
soup.find('div', class_='table-cell')
```

```python
soup.find_all('div', class_='table-cell')
```

```python
soup.find_all('div', class_='organizationName second table-cell team')
```

```python
data_rank = []
data_team =[]
data_liga =[]
data_nilaiskuad =[]

ranks = soup.find_all('div', class_='rank first table-cell rank')
for rank in ranks:
    data_rank.append(rank.text)
    
teams = soup.find_all('div', class_='organizationName second table-cell team')
for team in teams:
    data_team.append(team.text)
    
league = soup.find_all('div', class_='industry table-cell league')
for liga in league:
    data_liga.append(liga.text)
    
value = soup.find_all('div', class_="totalValue table-cell current value")
for nilaiskuad in value:
    data_nilaiskuad.append(nilaiskuad.text)
```

```python
football = {
    'Rank':data_rank,
    'Nama Team': data_team,
    'Asal Liga': data_liga,
    'Nilai Team': data_nilaiskuad
}

pd.DataFrame(football).to_excel('MostValuableTeamFootball.xlsx')
```
